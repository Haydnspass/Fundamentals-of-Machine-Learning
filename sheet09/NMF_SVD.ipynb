```python
import math
import matplotlib.pyplot as plt
import numpy as np

from sklearn.datasets import load_digits
from sklearn import decomposition

digits = load_digits()

X = digits["data"] / 255.
Y = digits["target"]
```

```python
n_components=10
model_NMF = decomposition.NMF(n_components=n_components)
model_NMF.fit_transform(X)
H = model_NMF.components_
```

```python
f, axarr = plt.subplots(2,5, figsize=(20,20))
for i in range(n_components):
    axarr[np.unravel_index(i, (2,5))[0],np.unravel_index(i, (2,5))[1]].imshow(np.reshape(H[i,:], (8,8)))
plt.show()
```

```python
mu = np.mean(X)
X_hat = X - mu

model_svd = decomposition.TruncatedSVD(n_components=63, n_iter=1000)
model_svd.fit(X)
V = model_svd.components_ + mu
```

```python
f, axarr = plt.subplots(2,5, figsize=(20,20))
for i in range(n_components):
    axarr[np.unravel_index(i, (2,5))[0],np.unravel_index(i, (2,5))[1]].imshow(np.reshape(V[i,:], (8,8)))
plt.show()
```
